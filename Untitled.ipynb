```python
import cv2
import numpy as np
```

```python
img = cv2.imread('Raw_Dataset/img-8.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, threshold = cv2.threshold(gray,170,255,cv2.THRESH_BINARY_INV)
contours, hierarchy = cv2.findContours(threshold,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
final = np.zeros(gray.shape,np.uint8)
mask = np.zeros(gray.shape,np.uint8)
list1 = []
```

```python
for i in range(0, len(contours)):
    cnt = contours[i]
    epsn = 0.1*cv2.arcLength(cnt,True)
    approx = cv2.approxPolyDP(cnt,epsn,True)
    size = cv2.contourArea(approx)
    eLeft = tuple(cnt[cnt[:, :, 0].argmin()][0])
    eRight = tuple(cnt[cnt[:, :, 0].argmax()][0])
    distance = np.sqrt((eLeft[0] - eRight[0])**2 + (eLeft[1] - eRight[1])**2)
    x,y,w,h = cv2.boundingRect(cnt)
    mask[x:x+h, y:y+w]=0
    if 700 > size > 220:
        cv2.drawContours(mask,contours,i,255,-1)
        cv2.drawContours(final,contours,i,255,2)

    elif 5 < distance < 17 and size > 25:
        list1.append(cnt)

```

```python
for i in list1:        
    cv2.drawContours(final, [i], -1, (255,255,255), -1)
```

```python
cv2.bitwise_not(final,final)
cv2.imwrite('c_orig.png', img)
cv2.imwrite('c_boxes.png', final)
cv2.imshow('img', img)
cv2.imshow('img2', final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
